```python
# Importar bibliotecas necessárias


import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Carregar dados e criar variáveis dummy para a coluna 'Escolaridade'

limite = pd.get_dummies(limite, prefix=['Escolaridade' ])

# Visualizar as primeiras linhas do dataframe

limite.head()

# Análise de Regressão Linear

# Configurar a fórmula para o modelo

```

```python
import statsmodels.formula.api as smf
```

```python
# Análise de Regressão Linear
# Configurar a fórmula para o modelo
# Modelo com todas as variáveis

# Ajustar modelo sem interceptação

function = 'LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediatovz+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos'
```

```python
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```

```python
# Ajustar modelo sem interceptação


function = 'LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos -1'

# Ajustar o modelo e exibir o resumo

model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```

```python
# Ajustar modelo sem 'Escolaridade_Fundamental_Medio'


function = 'LimitedoChequeEspecial~Idade+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+ -1'

model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```

```python
# Ajustar modelo apenas com variáveis específicas


function = 'LimitedoChequeEspecial~+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio -1'

model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```
